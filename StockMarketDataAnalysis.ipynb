```python
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import date

# Fetch stock historical data using yfinance
stock_data = yf.download("SBIN.BO", start="2018-01-15", end="2018-02-01")

# Filter for the specific option (Call option with strike price 300 and expiry date 2018-02-22)
option_data = stock_data[(stock_data.index <= "2018-02-22") & (stock_data.index >= "2018-01-15")]

# Plot the closing prices
option_data['Close'].plot()
plt.title("SBIN Call Option - Strike Price: 300, Expiry Date: 2018-02-22")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import date

# Fetch stock futures data using yfinance
stock_fut = yf.download("SBIN.BO", start="2023-01-15", end="2024-02-01")

# Plot the closing prices
stock_fut['Close'].plot()
plt.title("SBIN Stock Futures")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```
